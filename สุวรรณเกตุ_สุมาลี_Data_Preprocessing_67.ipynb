###ค่าผิดปกติ (outliers)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numeric_columns = ['pclass', 'survived', 'age', 'sibsp', 'parch', 'fare', 'body']

# สร้าง DataFrame เพื่อสรุปผล outliers
outliers_summary = pd.DataFrame(columns=['Column', 'Outliers'])

for col in numeric_columns:
    Q1 = titanic_df[col].quantile(0.25)
    Q3 = titanic_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # หาค่าที่เป็น outliers
    outliers = titanic_df[(titanic_df[col] < lower_bound) | (titanic_df[col] > upper_bound)][col]

    # เพิ่มข้อมูลในตารางสรุป
    outliers_summary = pd.concat([outliers_summary, pd.DataFrame({'Column': [col], 'Outliers': [outliers.values]})], ignore_index=True)

# แสดงตารางสรุป outliers
outliers_summary['Outliers'] = outliers_summary['Outliers'].apply(lambda x: ', '.join(map(str, x)) if len(x) > 0 else 'None')
outliers_summary
```

```python
# เลือกเฉพาะคอลัมน์ที่เป็นตัวเลขที่ต้องการตรวจสอบ
numeric_columns = ['pclass', 'survived', 'age', 'sibsp', 'parch', 'fare', 'body']

# สร้าง figure และ subplot
plt.figure(figsize=(14, 10))

# คำนวณ Q1, Q2, Q3 สำหรับแต่ละคอลัมน์และ plot boxplot พร้อมเส้น Q1, Q2, Q3
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(3, 3, i)  # ปรับเป็น 3x3 subplot เพื่อรองรับ 7 กราฟ

    # คำนวณ quartiles
    Q1 = titanic_df[col].quantile(0.25)
    Q2 = titanic_df[col].median()
    Q3 = titanic_df[col].quantile(0.75)

    # สร้าง boxplot
    sns.boxplot(data=titanic_df, y=col)

    # แสดงเส้น Q1, Q2 (median), Q3
    plt.axhline(Q1, color='blue', linestyle='--', label='Q1 (25%)')
    plt.axhline(Q2, color='green', linestyle='--', label='Q2 (Median)')
    plt.axhline(Q3, color='red', linestyle='--', label='Q3 (75%)')

    plt.title(f'Boxplot of {col}')
    plt.legend()

plt.tight_layout()
plt.show()
```

เมื่อตรวจสอบจำนวนค่าว่างแต่ละคอลัมน์แล้ว พบว่าคอลัมน์ body จะมีจำนวนค่าว่างมากที่สุด จำนวน 1188 ค่า

```python
#ตรวจสอบการแจกแจงปกติของคอลัมน์ body
import scipy.stats as stats
import matplotlib.pyplot as plt

stats.probplot(titanic_df['body'].dropna(), dist="norm", plot=plt)
plt.title('Q-Q Plot of body')
plt.show()
```

เนื่องจากคอลัมน์ body มีการแจกแจงใกล้เคียงแบบปกติและไม่มีค่า Outlier จึงจะจัดการค่าว่างโดนการแทนค่า Mean ลงไปในค่าว่าง

```python
titanic_df['body'] = titanic_df['body'].fillna(titanic_df['body'].mean())
titanic_df
```

```python
titanic_df.isnull().sum()
```

จัดการค่าว่างของคอลัมน์ age โดยการ dropna จะให้ subset=['age'] เนื่องจาก age จะเป็นตัวแปรที่นำไปใช้ต่อ จึงต้องการจัดการค่า null ออกทั้งหมด

```python
titanic_df = titanic_df.dropna(subset=['age'])
titanic_df
```

```python
titanic_df['cabin'].fillna('Unknown', inplace=True)
titanic_df['embarked'].fillna('Unknown', inplace=True)
titanic_df['boat'].fillna('Unknown', inplace=True)
titanic_df['home.dest'].fillna('Unknown', inplace=True)
titanic_df
```

```python
#จะจัดการค่าว่างของคอลัมน์ fare โดยการแทนที่ค่าว่างด้วย Median เนื่องจากคอลัมน์ fare มีค่า outlier มาก
titanic_df['fare'] = titanic_df['fare'].fillna(titanic_df['fare'].median())
titanic_df
```

# แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
* i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
* ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
* iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
* iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

## i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked

```python
titanic_df = pd.get_dummies(titanic_df, columns=['embarked'], prefix='embarked')
titanic_df
```

## ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1

```python
titanic_df['Sex'] = titanic_df['sex'].map({'male': 0, 'female': 1})
titanic_df
```

## iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard)

```python
titanic_df['FamilySize'] = titanic_df['sibsp'] + titanic_df['parch']
titanic_df
```

## iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
titanic_df['IsAlone'] = np.where(titanic_df['FamilySize'] == 0, 1, 0)
titanic_df
```

# ทำการ Normalize ข้อมูล
* i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling \
* ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1 \
* iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

pd.cut(np.array([1, 7, 5, 4, 6, 3]), 3)
```

```python
pd.cut(np.array([1, 7, 5, 4, 6, 3]), [0,2,5,8])
```

## i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling

```python
scaler = MinMaxScaler()
titanic_df[['fare', 'age']] = scaler.fit_transform(titanic_df[['fare', 'age']])
titanic_df
```

## ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1

```python
titanic_df['Fare_Per_Person'] = titanic_df['fare'] / (titanic_df['FamilySize'] + 1)
titanic_df
```

## iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
age_bins = [0, 12, 18, 50, 100]
age_labels = ['Child', 'Teenager', 'Adult', 'Senior']
titanic_df['AgeGroup'] = pd.cut(titanic_df['age'] * 100, bins=age_bins, labels=age_labels, right=False)
titanic_df
```
